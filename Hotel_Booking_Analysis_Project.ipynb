## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# missing values
df.isnull().sum().sort_values(ascending = False)
````

````python
# Replacing null values of company, agent and children columns with 0 and country column with 'nan'

df[['company','agent','children']] = df[['company','agent','children']].fillna(0)
df[['country']] = df[['country']].fillna('nan')
````

````python
# checking again after replacing null values
df.isnull().sum()
# now there are no null values in our dataset
````

````python
# duplicate values

df.duplicated().sum()

````

````python
# dropping duplicates
df = df.drop_duplicates()
````

````python
df.shape
````

````python

# Changing datatype of column 'reservation_status_date' from object to date_type

df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], format = '%Y-%m-%d')
````

````python
df.dtypes
````

````python
# Adding total staying days in hotels
df['total_stay'] = df['stays_in_weekend_nights']+df['stays_in_week_nights']

# Adding total people number as column, i.e. total types of person = num of adults + children + babies
df['total_people'] = df['adults']+df['babies']+df['children']
````

Chart - 18

What is the ADR across different months?

````python
# Grouping arrival_month and hotel on mean of adr
bookings_months=df.groupby(['arrival_date_month','hotel'])['adr'].mean().reset_index()

# Creating a month list to order the months in ascending
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Creating a dataset of months, hotel and their adr
bookings_months['arrival_date_month']=pd.Categorical(bookings_months['arrival_date_month'],categories=months,ordered=True)

# Sorting the months
bookings_months=bookings_months.sort_values('arrival_date_month')
bookings_months
````

````python

# Setting the chart size
plt.figure(figsize=(12,4))

# Plotting the values in a line chart
sns.lineplot(x=bookings_months['arrival_date_month'],y=bookings_months['adr'],hue=bookings_months['hotel'])

# Setting the labels and title
plt.title('ADR across each month', fontsize=16)
plt.xlabel('Month Name', fontsize=12)
plt.ylabel('ADR', fontsize=12)

# Show chart
plt.show()
````
